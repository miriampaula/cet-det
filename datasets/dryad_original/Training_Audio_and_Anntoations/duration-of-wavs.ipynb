```python
import soundfile as sf
import os

# Assuming wav_files_path_1 is a directory path
directory_path = wav_files_path_1

# List all WAV files in the directory
wav_files_in_directory = []
if os.path.isdir(directory_path):
    for filename in os.listdir(directory_path):
        if filename.lower().endswith('.wav'):
            wav_files_in_directory.append(os.path.join(directory_path, filename))
else:
    print(f"Warning: {directory_path} is not a valid directory or does not exist. Skipping file search in this path.")

# Combine with other explicitly defined WAV file paths
wav_files = wav_files_in_directory + [
    wav_files_path_3,
    wav_files_path_4,
    wav_files_path_5,
    wav_files_path_6
]

# Remove duplicates if any (though unlikely with current setup)
wav_files = list(set(wav_files))

total_duration_seconds = 0

print(f"Processing {len(wav_files)} WAV files...")
for file_path in wav_files:
    try:
        with sf.SoundFile(file_path, 'r') as f:
            duration = len(f) / f.samplerate
            total_duration_seconds += duration
            print(f"File: {file_path}, Duration: {duration:.2f} seconds")
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

total_duration_minutes = total_duration_seconds / 60
print(f"\nTotal duration of all WAV files: {total_duration_minutes:.2f} minutes")
```
